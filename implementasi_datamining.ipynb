```python
import numpy as np
from sklearn.datasets import load_iris
from scipy.spatial.distance import euclidean, cityblock, minkowski

iris = load_iris()
data = iris.data
target = iris.target

def hitung_jarak_custom(p1, p2, p_value=2):
    return np.power(np.sum(np.abs(p1 - p2)**p_value), 1/p_value)

print(f"Membandingkan Sampel Indeks 0 dengan 5 sampel pertama:\n")
print(f"{'Indeks':<10} | {'Euclidean':<12} | {'Manhattan':<12} | {'Minkowski(p=3)':<15}")
print("-" * 55)

for i in range(5):
    point_a = data[0]
    point_b = data[i]

    euc = euclidean(point_a, point_b)
    man = cityblock(point_a, point_b)
    min_dist = minkowski(point_a, point_b, p=3)

    print(f"Data-{i:<6} | {euc:<12.3f} | {man:<12.3f} | {min_dist:<15.3f}")

print("\n--- Analisis Pengelompokan Sederhana ---")
jarak_sama = euclidean(data[0], data[1])
jarak_beda = euclidean(data[0], data[100])

print(f"Jarak (Euclidean) spesies yang sama: {jarak_sama:.3f}")
print(f"Jarak (Euclidean) spesies yang berbeda: {jarak_beda:.3f}")
```
